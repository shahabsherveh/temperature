```python
from shastat.model import StructuralTimeSeriesConfig, StructuralTimeSeriesBuilder
import numpy as np
```

```python
model_builder = StructuralTimeSeriesBuilder()
```

```python
model_builder.build_model(X=None, y=np.random.normal(size=10).reshape(-1,1))
```
